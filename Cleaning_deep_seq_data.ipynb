```python
import pandas as pd

# Read CSV file into a DataFrame
df = pd.read_csv("C://monocycleoutput.csv",
                 index_col=False, names=['sequence', 'id_1', 'id_2'])

# Remove short sequences
df = df[df['sequence'].str.len() >= 13]

# Keep rows where the 'sequence' column starts with 'S'
df = df[df['sequence'].str.startswith('S')]

# Remove rows containing '*'
df = df[~df['sequence'].str.contains('\*')]

# Extracting first 15 AA into a new column 'ShortSequence'
df['ShortSequence'] = df['sequence'].str[:15]

# Displaying the result
print(df[['ShortSequence', 'id_1', 'id_2']])
```

```python
df
```

```python
# df['id_1'] = df['id_1'].apply(lambda x: [x])
# df['id_2'] = df['id_2'].apply(lambda x: [x])
df_combined = df.groupby('ShortSequence').agg({
    'id_1': lambda x: sum(x),
    'id_2': lambda x: sum(x)
}).reset_index()
df_combined
```

```python
df_combined.to_csv("C://monocyclecleaned2.csv",index=False)
pattern_1 = '^SC.*$'  #SC12C
pattern_2 = '^S.C.*$' #SXC10CX
pattern_3 = '^S..C.*$' #S2C8C2
pattern_4 = '^S...C.*$' #S3C9C
df_combined.loc[:,'group'] = 0
df_combined.loc[df_combined['ShortSequence'].str.match(pattern_1),'group'] =1
df_combined.loc[df_combined['ShortSequence'].str.match(pattern_2),'group'] =2
df_combined.loc[df_combined['ShortSequence'].str.match(pattern_3),'group'] =3
df_combined.loc[df_combined['ShortSequence'].str.match(pattern_4),'group'] =4
df_combined['group'] = df_combined['group'].astype(int)
df_combined
```
